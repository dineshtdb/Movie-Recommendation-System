## Lets move into machine learning

First lets do the feature extraction

To cluster movies based on their genres it would make sense to represent each movie as a binary vector indicating the presence & absence of each genre. Thus it would allow us to calculate the distances between movies based on their genre profiles.

Example:

movie X : 0,0,0,1,1,*1*

```python
genres_encoded = combined_df['Genre'].explode().str.get_dummies().groupby(level=0).sum()
genres_encoded.head()
```

Next lets move into clustering. We will use Kmeans clustering algorithm. K-means need a parameter on how many clusters should we want. To find the optimal amount of clusters we can use the "elbow method"

First we will iterate over a range of cluster numbers and calculates the within-cluster sum of squares.

Then we can look at the graph and find the "Elbow point" which tells us where adding more clusters would result into poor results.

More information refer to : https://en.wikipedia.org/wiki/Elbow_method_(clustering)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []  # Within-cluster sum of squares
cluster_range = range(1, 15)  # Checking for up to 15 clusters

for i in cluster_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(genres_encoded)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(cluster_range, wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

As we can see the "Elbow" point might be around 5 or 6 clusters. So lets assign the K-means with 6 *clusters*

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
combined_df['Cluster'] = kmeans.fit_predict(genres_encoded)

combined_df[['Title', 'Genre', 'Cluster']].head()
```

Next we create code that visualizes movies based on their genre clustering using t-SNE (t-distributed Stochastic Neighbor Embedding), a dimensionality reduction technique. Movies are plotted in a 2D space where similar movies (based on genre) are grouped together. The function visualize_movies accepts a movie title and highlights it in the visualization. In the provided example, the movie "Star Wars: Episode VI - Return of the Jedi" is highlighted in red amidst other movies grouped by their genre clusters.

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

def visualize_movies(selected_movie, data=genres_encoded, clusters=combined_df['Cluster']):
    """
    Visualizes the movies based on their genre clustering and highlights the selected movie.

    Parameters:
    - selected_movie (str): The title of the movie to be highlighted.
    - data (DataFrame): One-hot encoded genre data.
    - clusters (Series): Cluster assignments for each movie.

    Returns:
    - A 2D plot of movies based on their genre clustering.
    """

    tsne = TSNE(n_components=2, random_state=42)
    reduced_data = tsne.fit_transform(data)

    plt.figure(figsize=(15, 10))
    for cluster_num in clusters.unique():
        cluster_data = reduced_data[clusters == cluster_num]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_num}', alpha=0.6)

    movie_idx = combined_df[combined_df['Title'] == selected_movie].index[0]
    plt.scatter(reduced_data[movie_idx, 0], reduced_data[movie_idx, 1], color='red', s=100, edgecolors='black', label='Selected Movie')

    plt.title(f"Movie Clustering based on Genres\n(Selected Movie: {selected_movie})", fontsize=16)
    plt.xlabel('t-SNE Component 1')
    plt.ylabel('t-SNE Component 2')
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_movies("Star Wars: Episode VI - Return of the Jedi")
```

Lets move forward on trying to suggest 5 closest movies based on euclidian distance:

```python
from scipy.spatial import distance

def recommend_movies_nearest_updated_euclidian(movie_title, num_recommendations=5):
    """
    Returns a list of movie recommendations based on the provided movie title.

    Parameters:
    - movie_title (str): The title of the movie for which recommendations are needed.
    - num_recommendations (int): Number of movie recommendations desired.

    Returns:
    - list of recommended movie titles.
    """
    if movie_title not in combined_df['Title'].values:
        return "Movie title not found in the dataset."

    movie_data = combined_df[combined_df['Title'] == movie_title]
    movie_cluster = movie_data['Cluster'].iloc[0]
    movie_vector = genres_encoded.loc[movie_data.index].values[0]

    cluster_movies = combined_df[combined_df['Cluster'] == movie_cluster]
    cluster_movies_vectors = genres_encoded.loc[cluster_movies.index]

    distances = cluster_movies_vectors.apply(lambda row: distance.euclidean(row, movie_vector), axis=1)

    nearest_movies = distances.nsmallest(num_recommendations + 1).index  # +1 to ensure enough recommendations
    recommended_movie_titles = combined_df.loc[nearest_movies]['Title'].tolist()

    if movie_title in recommended_movie_titles:
        recommended_movie_titles.remove(movie_title)

    return recommended_movie_titles

recommend_movies_nearest_updated_euclidian("Star Wars: Episode VI - Return of the Jedi")
```

Hmm is euclidian best way to calculate distances? Especially with high dimensional data? Lets try cosine simiuliary which measures the cosine of the angle between two vectors. It is used more for high dimensional data.

```python
from scipy.spatial import distance

def recommend_movies_nearest_updated_cosine(movie_title, num_recommendations=5):
    """
    Returns a list of movie recommendations based on the provided movie title using cosine similarity.

    Parameters:
    - movie_title (str): The title of the movie for which recommendations are needed.
    - num_recommendations (int): Number of movie recommendations desired.

    Returns:
    - list of recommended movie titles.
    """
    if movie_title not in combined_df['Title'].values:
        return "Movie title not found in the dataset."

    movie_data = combined_df[combined_df['Title'] == movie_title]
    movie_cluster = movie_data['Cluster'].iloc[0]
    movie_vector = genres_encoded.loc[movie_data.index].values[0]

    cluster_movies = combined_df[combined_df['Cluster'] == movie_cluster]
    cluster_movies_vectors = genres_encoded.loc[cluster_movies.index]

    similarities = cluster_movies_vectors.apply(lambda row: distance.cosine(row, movie_vector), axis=1)

    nearest_movies = similarities.nlargest(num_recommendations + 1).index  # +1 to ensure enough recommendations

    recommended_movie_titles = combined_df.loc[nearest_movies]['Title'].tolist()

    if movie_title in recommended_movie_titles:
        recommended_movie_titles.remove(movie_title)

    return recommended_movie_titles

recommend_movies_nearest_updated_cosine("Star Wars: Episode VI -# Save the K-Means model to a file Return of the Jedi")
```
